```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Sample house price dataset
data = {
    "Area": [800, 1000, 1200, 1500, 1800, 2000, 2200, 2500],
    "Bedrooms": [2, 2, 3, 3, 4, 4, 4, 5],
    "Bathrooms": [1, 2, 2, 2, 3, 3, 3, 4],
    "Age": [20, 15, 10, 8, 5, 3, 2, 1],
    "Price": [2000000, 2800000, 3500000, 4500000,
              5500000, 6500000, 7200000, 8500000]
}

df = pd.DataFrame(data)
df
```

```python
X = df[["Area", "Bedrooms", "Bathrooms", "Age"]]
y = df["Price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
# New house details
new_house = pd.DataFrame(
    [[1600, 3, 2, 6]],
    columns=["Area", "Bedrooms", "Bathrooms", "Age"]
)

predicted_price = model.predict(new_house)
print("Predicted House Price:", int(predicted_price[0]))
```
