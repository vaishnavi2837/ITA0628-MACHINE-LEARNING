```python
import numpy as np
```

```python
X = np.array([1, 2, 3, 8, 9, 10]).reshape(-1, 1)
n = len(X)
k = 2   # number of clusters
```

```python
means = np.array([[2.0], [9.0]])
variances = np.array([1.0, 1.0])
weights = np.array([0.5, 0.5])
```

```python
def gaussian(x, mean, var):
    return (1 / np.sqrt(2 * np.pi * var)) * np.exp(-(x - mean)**2 / (2 * var))
```

```python
for iteration in range(10):

    # E-step
    responsibilities = np.zeros((n, k))
    for i in range(n):
        for j in range(k):
            responsibilities[i, j] = weights[j] * gaussian(X[i], means[j], variances[j])
        responsibilities[i] /= np.sum(responsibilities[i])

    # M-step
    for j in range(k):
        r_sum = np.sum(responsibilities[:, j])
        means[j] = np.sum(responsibilities[:, j].reshape(-1, 1) * X) / r_sum
        variances[j] = np.sum(
            responsibilities[:, j] * (X.flatten() - means[j])**2
        ) / r_sum
        weights[j] = r_sum / n
```

```python
print("Final Means:", means.flatten())
print("Final Variances:", variances)
print("Final Weights:", weights)
```
